```python
#Ques = 4

import numpy as np
import matplotlib.pyplot as plt

arr=np.linspace(0, 20, 100)
def fuzzy(gamma):
    
    list_x=[]
    for i in arr:
        factor=gamma*abs((10/max(i,10)-(i/max(i,10)))) 
    # print(factor)
        if factor ==0:
            x=1
        if factor>0 and factor<=1:
            x=1-factor
        if factor >1:
            x=0
        list_x.append(x)
    return list_x

for i in range(1,6):
    y=fuzzy(i)
    print(y)
    print(len(y))
```

```python
# ques=5

#Design and implement (coding) a fuzzy set for finding out matching of 2 patterns represented 
#by feature vectors V1=[0.1,0.4,0.5] and V2=[-0.4,0.1,-0.3]. Use Euclidean distance to measure the distance 
#between them and decide the fuzzy membership value in terms of this distance. More is the distance 
#lesser will be the fuzzy membership value

import math
import numpy as np

v1=[0.1,0.4,0.5]
v2=[-0.4,0.1,-0.3]
v11=[1,2]
v22=[2,2]

point1=np.array((0.1,0.4,0.5))
point2=np.array((-0.4,0.1,-0.3))
square=np.square(point1 - point2)
sum_square=np.sum(square)
dis=np.sqrt(sum_square)
print(dis)

distance = math.dist(v1,v2)
print("euclidean distance:",distance)


gamma=1
factor = gamma*distance
if factor ==0:
    x=1
if factor>0 and factor<=1:
    x=1-factor
if factor >1:
    x=0
print("membership value:",x)
```

```python
# ques = 1

import numpy as np

array=[[0.5,-1.1,0.24,-0.22,-1.8], [0.15,-1.4,0.11,-0.28,-0.3], [0.1,-1.4,0.11,0.6,-1.11]]
desired=[1,-1,1]
x = np.array(array)
weight=[0.1, 0.17,0.2,-1.1, -0.4]
output = []
c = 2.5
p = 0 #epochs

for p in range(10):
    output=[]
    print(p)
    for i in range(3): #3 times train
        y = 0
        for j in range(len(x[i])):
            y = y + x[i][j]*weight[j]  #net
        if(y>0):
            y=1
        elif(y<0):
            y=-1
        output.append(y)  #o
        weight = weight + c*(desired[i]-y)*x[i]   #delta w = c*(desired[i]-y)*x[i]
        print("List Output",weight,output)
    error_arr=np.absolute(np.array(output)- np.array(desired))
    print(error_arr)
    err=np.mean(error_arr)
    print(err)
    print("\n")
```

```python
# ques = 2

import numpy as np

array=[[0.5,-1.1,0.24,-0.22,-1.8], [0.15,-1.4,0.11,-0.28,-0.3], [0.1,-1.4,0.11,0.6,-1.11]]
desired=[1,-1,1]
x = np.array(array)
weight=[0.1, 0.17,0.2,-1.1, -0.4]
output = []
c = 0.6
p = 0 #epochs
err= 10000 #To start.
while (err >0.001):
    output=[]
    print(p)
    for i in range(3): #3 times train
        y = 0
        for j in range(len(x[i])):
            y = y + x[i][j]*weight[j]
        if(y>0):
            y=1
        elif(y<0):
            y=-1
        output.append(y)
        weight = weight + c*(desired[i]-y)*x[i]
        print("List Output",weight,output)

    error_arr=np.absolute(np.array(output)- np.array(desired))
    print(error_arr)
    err=np.mean(error_arr)
    print(err)
    print("\n")
    p=p+1
print("no of epochs required:",p-1)
```
